# 认识NumPy
NumPy 是 Python 的一个库，用于科学计算。
使用 NumPy，你可以处理数组，进行数值计算，进行线性代数，进行傅立叶变换，等等。

### 为什么使用 NumPy？
1. 对于同样的数值计算任务，使用Numpy 的代码比纯 Python 代码:(1)代码更简洁：NumPy 提供了大量的数学函数，可以方便地进行数值计算。(2)执行效率更高：NumPy 的核心是用 C 语言编写的，因此执行效率更高。
2. NumPy 是众多数据科学库的依赖，例如 Pandas、Matplotlib、Scikit-learn 等。

### NumPy的主要特点
1. 强大的N维数组

2. 数值计算工具

NumPy 提供全面的数学函数、随机数生成器、<br>线性代数、傅立叶变换等。

3. 开源且表现出色

在 GitHub 上公开开发和维护。<br>（https://github.com/numpy ）<br>
NumPy 的核心是 C 代码。<br>享受 Python 的灵活性和编译代码的速度。

4. 是众多数据科学库的依赖

### Numpy与原生Python的区别

```python
import numpy as np
```

```python
np.__version__
```

```python
#原生Python实现列表乘法
a = [1, 2, 3]
b = [4, 5, 6]
c = []
for i in range(len(a)):
    c.append(a[i]*b[i])
c
```

```python
# NumPy实现列表乘法
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a * b
c
```

Numpy 不仅编码效率高，执行效率也高。

例：一个包含一百万整数的数组，和一个等价的 Python 列表：

```python
import numpy as np
my_arr = np.arange(1000000)
my_list = list(range(1000000))
```

```python
%timeit np.sum(my_arr)
```

```python
%timeit sum(my_list)
```
